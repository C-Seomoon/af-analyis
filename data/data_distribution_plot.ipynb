```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# 1) DataFrame에서 x, y 값 추출
x = df['ipTM']
y = df['DockQ']

# 2) Pearson 상관계수 계산
r, p_value = stats.pearsonr(x, y)

# 3) 단순 선형 회귀(직선) 계수 계산
slope, intercept, r_value, p_val_linreg, std_err = stats.linregress(x, y)
r_sq = r_value**2

# 4) 산점도 + 회귀선 그리기
plt.figure()
plt.scatter(x, y)  # 산점도
# 회귀선
x_vals = np.array([x.min(), x.max()])
y_vals = slope * x_vals + intercept
plt.plot(x_vals, y_vals, linestyle='--')

# 5) 플롯 위에 텍스트로 상관계수·R² 표시
#   transform=... 으로 좌표계를 0~1 범위의 축 좌표(axes fraction)로 지정
plt.text(
    0.05, 0.95,
    f'Pearson r = {r:.2f}\nR² = {r_sq:.2f}',
    transform=plt.gca().transAxes,
    verticalalignment='top'
)

# 6) 레이블·타이틀
plt.xlabel('ipTM')
plt.ylabel('DockQ')
plt.title('ipTM vs DockQ (with Correlation Metrics)')

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats


import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def plot_corr_with_dockq(df, col, dockq_col='DockQ', figsize=(6,6), alpha=0.7):
    """
    df: pandas DataFrame
    col: 상관관계를 보고 싶은 컬럼명 (문자열)
    dockq_col: DockQ 컬럼명 (기본값 'DockQ')
    figsize: 그림 크기 튜플 (너비, 높이)
    alpha: 산점도 포인트 투명도
    """
    # 1) x, y 추출
    x = df[col]
    y = df[dockq_col]

    # 2) 상관계수 계산
    r_p, p_p = stats.pearsonr(x, y)        # Pearson
    r_s, p_s = stats.spearmanr(x, y)      # Spearman
    r_k, p_k = stats.kendalltau(x, y)     # Kendall Tau

    # 3) 단순 선형 회귀 계수 및 결정계수
    slope, intercept, r_value, p_val_linreg, std_err = stats.linregress(x, y)
    r_sq = r_value**2

    # 4) 플롯
    plt.figure(figsize=figsize)
    plt.scatter(x, y, alpha=alpha)
    # 회귀선 (0–1 범위)
    x_vals = np.array([0, 1])
    y_vals = slope * x_vals + intercept
    plt.plot(x_vals, y_vals, linestyle='--')

    # 5) 축 범위 고정
    plt.xlim(0, 1)
    plt.ylim(0, 1)

    # 6) 텍스트로 상관계수들 및 R² 표시
    plt.text(
        0.05, 0.95,
        (f'Pearson r = {r_p:.2f}\n'
         f'Spearman ρ = {r_s:.2f}\n'
         f'Kendall τ = {r_k:.2f}\n'
         f'R² = {r_sq:.2f}'),
        transform=plt.gca().transAxes,
        verticalalignment='top'
    )

    # 7) 레이블·타이틀
    plt.xlabel(col)
    plt.ylabel(dockq_col)
    plt.title(f'{col} vs {dockq_col}')
    plt.tight_layout()
    plt.show()

```

```python
plot_corr_with_dockq(df, 'pdockq_AH')
plot_corr_with_dockq(df, 'pdockq2_AH')
plot_corr_with_dockq(df, 'mpdockq_AH')
plot_corr_with_dockq(df, 'LIS_AH')
plot_corr_with_dockq(df, 'chain_pair_iptm_AH')
plot_corr_with_dockq(df, 'scaled_model_RMSD')
plot_corr_with_dockq(df, 'scaled_query_RMSD')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1) 범주 정의용 bins 와 labels
bins   = [0.00, 0.23, 0.49, 0.80, 1.00]
labels = ['Incorrect', 'Acceptable quality', 'Medium quality', 'High quality']

# 2) DockQ 값을 범주화해서 새로운 컬럼 추가
df['DockQ_label'] = pd.cut(df['DockQ'], bins=bins, right=False, labels=labels)

# 3) 전체 샘플 수 및 성공률 계산
total = len(df)
success_count = df[df['DockQ'] >= 0.23].shape[0]
success_rate = success_count / total * 100  # %

# 4) 각 범주별 개수 및 비율 집계
counts = df['DockQ_label'].value_counts().reindex(labels)
percentages = counts / total * 100

# 5) 레이블별 색상 지정
colors = ['#d62728', '#ff7f0e', '#bcbd22', '#2ca02c']

# 6) 플롯 그리기
plt.figure(figsize=(8, 5))
bars = plt.bar(counts.index, counts.values, color=colors, edgecolor='k')

# 7) 막대 위에 개수 및 비율 표시
for bar, pct in zip(bars, percentages):
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,
        height + total * 0.01,  # 전체 높이 기준 약간 위에
        f'{int(height)}\n({pct:.1f}%)',
        ha='center', va='bottom'
    )

# 8) 성공률 텍스트 표시
plt.text(
    0.02, 0.90,
    f'Success rate\n(DockQ ≥ 0.23): {success_rate:.1f}%',
    transform=plt.gca().transAxes,
    fontsize=11,
    bbox=dict(boxstyle='round,pad=0.3', facecolor='white', edgecolor='gray')
)

# 9) 레이블·타이틀·레이아웃
plt.xlabel('DockQ Quality Category')
plt.ylabel('Count')
plt.title('DockQ Score Category Distribution')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from sklearn.metrics import r2_score


# 0 나누기 처리: 0일 땐 NaN 으로
df['RMSD_ratio'] = df['model_avg_RMSD'] / df['query_avg_RMSD'].replace(0, np.nan)

# 1) 가정: df 에는 여러분의 메트릭(metric_col)과 실제 스코어(true_col)가 들어있다
metric_col = 'model_avg_RMSD'   # 여러분의 메트릭 컬럼명
true_col   = 'DockQ'  # 피팅 대상으로 삼을 실제 스코어 컬럼명




x = df[metric_col].values
y = df[true_col].values

# 2) 피팅할 시그모이드 함수 정의
def sigmoid(x, L, k, x0, b):
    """
    L : Curve의 최대 진동폭(amplitude)
    k : 기울기(slope)
    x0: Mid-point
    b : Baseline offset
    """
    return L / (1 + np.exp(-k*(x - x0))) + b

# 3) 초기 파라미터 추정값 설정 (이 값들이 너무 벗어나면 수렴이 어려우니 대략 맞춰주세요)
p0 = [
    max(y) - min(y),      # L : y 범위
    1.0,                  # k : 보통 0.01 ~ 1 사이
    np.median(x),         # x0: x의 중앙값
    min(y)                # b : y의 최소값
]

# 4) curve_fit 으로 최적 파라미터 찾기
popt, pcov = curve_fit(sigmoid, x, y, p0=p0, maxfev=10000)
L_fit, k_fit, x0_fit, b_fit = popt

print(f'Fitted parameters:\n L={L_fit:.3f}, k={k_fit:.3f}, x0={x0_fit:.3f}, b={b_fit:.3f}')

# 5) 피팅 결과를 바탕으로 예측값 계산
y_pred = sigmoid(x, *popt)

# 6) 결정계수(R²) 계산
r2 = r2_score(y, y_pred)
print(f'R² = {r2:.3f}')

# 7) 결과 시각화
plt.figure(figsize=(6,6))
plt.scatter(x, y, label='observed', alpha=0.6)
# 시그모이드 곡선 (x축 범위에 맞춰 부드럽게)
xs = np.linspace(x.min(), x.max(), 200)
plt.plot(xs, sigmoid(xs, *popt), color='red', linestyle='--', label='fitted sigmoid')

plt.xlabel(metric_col)
plt.ylabel(true_col)
plt.title(f'Sigmoid Fit\n$R^2$ = {r2:.3f}')
plt.legend()
plt.tight_layout()
plt.show()
```
