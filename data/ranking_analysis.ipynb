```python
import numpy as np
import pandas as pd
from typing import List, Dict, Any

def compute_ranking_metrics(
    df: pd.DataFrame,
    query_col: str = 'query',
    pred_score_col: str = 'model_avg_RMSD',
    true_score_col: str = 'DockQ',
    topk: List[int] = [5, 10],
    lower_better: bool = False
) -> Dict[str, Any]:
    """
    pred_score_col과 true_score_col 각각에 대해 순위를 매기고,
    다음 지표를 계산합니다:

      1. Top-1 Accuracy  
         => true_rank==1인 모델이 pred_rank==1인지 확인  
      2. Mean Rank Difference  
         => pred_rank==1인 모델의 true_rank 가 1에서 얼마나 떨어져 있는지 평균  
      3. Median Rank Difference  
         => 위 차이의 중앙값  
      4. MRR  
         => pred_rank==1인 모델의 reciprocal true_rank 의 평균  
      5. Top-k Recall  
         => pred_rank==1인 모델이 실제 true_rank ≤ k 안에 있는 비율

    lower_better=True 이면 pred_score_col 값이 작을수록 순위 1위,
    False 이면 클수록 순위 1위로 매깁니다.
    """

    n_queries = df[query_col].nunique()
    top1_hits = 0
    rank_diffs: List[float] = []
    reciprocal_ranks: List[float] = []
    topk_hits = {k: 0 for k in topk}

    for _, grp in df.groupby(query_col):
        # (1) 명시적으로 순위 계산
        grp = grp.copy()
        grp['pred_rank'] = grp[pred_score_col].rank(method='first', ascending=lower_better)
        grp['true_rank'] = grp[true_score_col].rank(method='first', ascending=False)

        # (2) Top-1 Accuracy: true_rank==1 모델이 pred_rank==1인지?
        true_top_idx = grp['true_rank'].idxmin()  # 같은 as grp['true_rank']==1
        if grp.loc[true_top_idx, 'pred_rank'] == 1:
            top1_hits += 1

        # (3) Mean/Median Rank Difference, MRR, Top-k Recall
        # pred 기준 1위인 모델
        pred_top_idx = grp['pred_rank'].idxmin()
        true_rank_of_pred_top = grp.loc[pred_top_idx, 'true_rank']

        # 차이
        diff = abs(true_rank_of_pred_top - 1)
        rank_diffs.append(diff)
        reciprocal_ranks.append(1.0 / true_rank_of_pred_top)

        for k in topk:
            if true_rank_of_pred_top <= k:
                topk_hits[k] += 1

    # 결과 조합
    metrics: Dict[str, Any] = {
        'Top-1 Accuracy':     top1_hits / n_queries,
        'Mean Rank Difference':  np.mean(rank_diffs),
        'Median Rank Difference': np.median(rank_diffs),
        'MRR':                np.mean(reciprocal_ranks),
    }
    for k, hits in topk_hits.items():
        metrics[f'Top-{k} Recall'] = hits / n_queries

    return metrics
```
