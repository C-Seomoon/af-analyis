```python
import pandas as pd

# 1. CSV 로드
df = pd.read_csv('./classification/results/cls_final_20250425_203802/model_comparison_summary.csv')  # 실제 파일 경로로 바꿔주세요

# 2. 관심 있는 메트릭 리스트
metrics = [
    'accuracy', 'precision', 'recall', 'f1',
    'roc_auc', 'pr_auc', 'balanced_accuracy',
    'mcc', 'best_inner_cv_roc_auc'
]

# 3. "mean"과 "std" 컬럼을 붙여서 새로운 컬럼 생성
#    e.g. accuracy -> "0.864 ± 0.071"
df_formatted = df.set_index('model_name').copy()
for m in metrics:
    mean_col = f'{m}_mean'
    std_col  = f'{m}_std'
    df_formatted[m] = (
        df_formatted[mean_col].map('{:.3f}'.format)
        + ' ± '
        + df_formatted[std_col].map('{:.3f}'.format)
    )

# 4. 최종적으로 보고 싶은 컬럼 순서대로 선택
result = df_formatted[metrics]

# 5. 출력
#    - Jupyter 환경이라면 아래 한 줄만
display(result)

#    - 터미널/스크립트라면 to_markdown()을 이용해 마크다운 표로 출력
print(result.to_markdown())
```

```python
import pandas as pd

csv_path = "./regression/results/rg_renewal_test_20250429_033939/model_comparison_summary.csv"
df = pd.read_csv(csv_path)
metrics = [
    ("r2_mean", "r2_std", "R²"),
    ("mse_mean", "mse_std", "MSE"),
    ("rmse_mean", "rmse_std", "RMSE"),
    ("mae_mean", "mae_std", "MAE"),
    ("best_inner_cv_score_mean", "best_inner_cv_score_std", "Best-CV")
]

for mean_col, std_col, label in metrics:
    df[label] = (
        df[mean_col].map("{:.3f}".format)
        + " ± "
        + df[std_col].map("{:.3f}".format)
    )


display_cols = ["model_name"] + [label for *_, label in metrics]
result = df[display_cols]

from IPython.display import display
display(result)
```

```python
import pandas as pd
from IPython.display import display

# 1) CSV 파일 로드
csv_path = './regression/results/rg_renewal_test_20250429_033939/metrics_by_fold.csv'  # 실제 파일 경로로 변경하세요
df = pd.read_csv(csv_path)

# 2) pivot 으로 모델별 폴드별 R2 매트릭스 생성
r2_pivot = df.pivot(index='fold', columns='model', values='r2')

# 3) 보기 좋게 정렬 (폴드 번호 순, 모델 이름 알파벳 순)
r2_pivot = r2_pivot.sort_index().sort_index(axis=1)


print("■ Fold별 R² 값 (style.format) ■")
display(
    r2_pivot.style
            .format("{:.2f}")
            .set_caption("Fold별 R² (소수점 둘째 자리)")
)

# # 4) 출력
# print("■ Fold별 R² 값 ■")
# display(r2_pivot)
```

```python
import pandas as pd
from IPython.display import display

# 1) CSV 파일 로드
csv_path = '/home/cseomoon/appl/af_analysis-0.1.4/model/classification/results/cls_renewal_20250429_144318/model_comparison_summary.csv'  # 실제 파일 경로로 바꿔주세요
df = pd.read_csv(csv_path)

# 2) 처리할 메트릭 리스트
metrics = [
    'accuracy', 'precision', 'recall', 'f1',
    'roc_auc', 'pr_auc', 'balanced_accuracy', 'mcc'
]

# 3) 각 메트릭에 대해 "mean ± std" 문자열 생성
for m in metrics:
    mean_col = f'{m}_mean'
    std_col  = f'{m}_std'
    combined = f'{m}'
    df[combined] = df.apply(
        lambda row: f"{row[mean_col]:.2f}±{row[std_col]:.2f}", axis=1
    )

# 4) 최종 표시할 컬럼 순서 지정
display_cols = ['model_name'] + metrics + ['best_inner_cv_pr_auc_mean', 'best_inner_cv_pr_auc_std']
# best_inner_cv_pr_auc도 같은 형식으로 합칠 경우:
df['inner_pr_auc'] = df.apply(
    lambda row: f"{row['best_inner_cv_pr_auc_mean']:.2f}±{row['best_inner_cv_pr_auc_std']:.2f}", axis=1
)
display_cols = ['model_name'] + metrics + ['inner_pr_auc']

# 5) 결과 출력
result_df = df[display_cols].set_index('model_name')
display(result_df)
```
