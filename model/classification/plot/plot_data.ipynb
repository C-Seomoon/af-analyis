```python
import json
import pandas as pd

results='/home/cseomoon/appl/af_analysis-0.1.4/model/classification/results/cls_pipeline_20250527_191338_PPT/final_models_info.json'

with open(results, 'r', encoding='utf-8') as f:
    data = json.load(f)

data['rf']
```

```python
# Create parameters DataFrame
params_df = pd.DataFrame({m: v["best_params"] for m, v in data.items()}).T
test_df=pd.DataFrame({m: v["test_metrics"] for m, v in data.items()}).T
# Create nested CV performance DataFrame with mean ± std
perf_df = pd.DataFrame({
    m: {
        k.split("_mean")[0]: f"{v:.3f} ± {data[m]['nested_cv_metrics'][k.replace('_mean','_std')]:.3f}"
        for k, v in v["nested_cv_metrics"].items() if k.endswith("_mean")
    }
    for m, v in data.items()
}).T


#params_df
display(perf_df.round(3))
display(test_df.round(3))
```

```python
import os
import json
import pandas as pd

# 1. features 디렉토리 경로
base_dir = '/home/cseomoon/appl/af_analysis-0.1.4/model/classification/features'

# 2. 각 feature별 metrics.json을 읽어 리스트에 저장
rows = []
for feature in os.listdir(base_dir):
    metrics_path = os.path.join(base_dir, feature, 'metrics.json')
    if os.path.isfile(metrics_path):
        with open(metrics_path, 'r') as f:
            metrics = json.load(f)
        metrics['feature'] = feature
        rows.append(metrics)

# 3. DataFrame 생성 및 feature를 인덱스로 설정
df = pd.DataFrame(rows).set_index('feature')

# 4. 결과 확인
display(df)
```

```python
df.sort_values(by='roc_auc', ascending=False).head(20)
```
