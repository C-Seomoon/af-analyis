```python
import numpy as np

# 기본 정보 확인
print("객체 타입:", type(model))
print("모델 수:", len(model.models))

# 첫 번째 모델 정보 출력
print("\n첫 번째 모델 정보:")
print("원자 수:", model.models[0].len)
print("체인 목록:", list(np.unique(model.models[0].chain)))
print("아미노산 수:", len(np.unique(model.models[0].uniq_resid)))

# 체인별 잔기 수 확인
print("\n체인별 잔기 수:")
for chain in np.unique(model.models[0].chain):
    chain_resids = model.models[0].uniq_resid[model.models[0].chain == chain]
    print(f"체인 {chain}: {len(np.unique(chain_resids))} 잔기")

# CA 원자 정보
ca_atoms = model.select_atoms("name CA")
print("\nCA 원자 정보:")
print("CA 원자 수:", ca_atoms.len)

# 추가 정보 확인을 위한 메서드
print("\n사용 가능한 메서드:")
print([method for method in dir(model) if not method.startswith('_')][:10])  # 처음 10개만
```
